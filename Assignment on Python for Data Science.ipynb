# 5. Perform meaningful visualizations using the dataset

### 1.Scatter Plot

```python
sns.scatterplot(data=iris_data,x='SL', y='SW')
plt.show()
```

### 2. Pairplot to examine the correlations

```python
plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
sns.pairplot(data = iris_data, hue="Classification", height=3);
plt.show()
```

###  3.Count Plot - displays the number of observations for a categorical variable using bars.

```python
sns.countplot(x=iris_data.Classification);
```

###  4.Box Plot-shows the distribution of quantitative data across a categorical variable.

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Classification',y='SW',data=iris_data ,palette='YlGnBu');
```

### 5.violin plot-It shows the distribution of quantitative data across several levels of categorical variables. The violin plot also features a kde of the underlying distribution

```python
sns.violinplot(x='Classification', y='PL', data=iris_data, palette='OrRd');
```
